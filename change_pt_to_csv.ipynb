```python
import torch

filepath = 'scaled_act/4bit/gs128/scaled_act_min_opt_125m.pt'
buffer = torch.load(filepath)
buffer
```

```python
temp = dict()
for key in buffer:
    key_type = key.split('.')[-1]
    if key_type in temp:
        temp[key_type].append(buffer[key].item())
    else:
        temp[key_type] = list()
        temp[key_type].append(buffer[key].item())
```
